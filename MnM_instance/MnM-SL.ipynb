```r
install.packages("softImpute")
#Installing package into ‘/home/nbuser/R’
#(as ‘lib’ is unspecified)
library(softImpute)
require(softImpute)
## Loading required package: softImpute
## Loading required package: Matrix
## Loaded softImpute 1.4
x <- read.csv('final_rating_data.csv')
fits=softImpute(x,trace=TRUE,type="svd")
# 1 : obj 0.51518 ratio 0.002312619 
# 2 : obj 0.48683 ratio 0.0001041141 
# 3 : obj 0.48542 ratio 8.662155e-06 
fits
```
